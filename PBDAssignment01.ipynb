```python
f=df[['diagnosis','gender']].value_counts()
u=df[['diagnosis','gender']].value_counts().keys()
plt.pyplot.pie(f, radius=2,autopct='%1.1f%%')
plt.pyplot.legend(loc='upper right',labels = u, fontsize = 60,bbox_to_anchor=(2,5))
plt.pyplot.show()
```
